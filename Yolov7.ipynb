```python
import requests
import requests

# Link video từ web
video_url = "https://www.shutterstock.com/shutterstock/videos/3419503927/preview/stock-footage-caring-primary-school-teacher-using-a-tv-with-green-screen-display-explaining-an-exercise-to-a.webm"
video_path = "/content/drive/MyDrive/BTL_AII/video_aa.mp4"  # Đường dẫn để lưu video

# Tải video từ URL
response = requests.get(video_url)
if response.status_code == 200:
    with open(video_path, "wb") as file:
        file.write(response.content)
    print(f"✅ Video đã được tải về: {video_path}")
else:
    print(f"❌ Không thể tải video từ URL: {video_url}")
```

```python
import subprocess

cmd = [
    "python3", "/content/yolov7/detect.py",
    "--weights", "/content/drive/MyDrive/BTL_AII/Yolo7_BTL/weights/best.pt",
    "--source", "/content/drive/MyDrive/Capcut/1.MOV",
    "--img-size", "640",
    "--conf-thres", "0.1",
    "--save-txt",
    "--save-conf",
    "--project", "runs/detect",
    "--name", "detect_output",
    "--exist-ok"
]


result = subprocess.run(cmd, capture_output=True, text=True)

print(result.stdout)
print(result.stderr)
```

```python
import subprocess
from PIL import Image
import IPython.display as display

# Lệnh chạy YOLOv7 detect.py
cmd = [
    "python3", "/content/yolov7/detect.py",
    "--weights", "/content/drive/MyDrive/BTL_AII/Yolo7_BTL/weights/best.pt",
    "--source", "/content/drive/MyDrive/BTL_AII/AI.v3-ai.yolov7pytorch/test/images/full_video_0682_jpg.rf.c945ee13a4d9f56de802f062666cd2f8.jpg",  # THÊM DẤU PHẨY Ở ĐÂY
    "--img-size", "640",
    "--conf-thres", "0.1",
    "--save-txt",
    "--save-conf",
    "--device", "0",
    "--project", "runs/detect",
    "--name", "detect_output",
    "--exist-ok"
]


# Chạy YOLOv7 detect.py
result = subprocess.run(cmd, capture_output=True, text=True)
print(result.stdout)
print(result.stderr)

# 🖼 Hiển thị ảnh có bounding box
output_img_path = "/content/drive/MyDrive/BTL_AII/AI.v3-ai.yolov7pytorch/test/images/full_video_0682_jpg.rf.c945ee13a4d9f56de802f062666cd2f8.jpg"

try:
    img = Image.open(output_img_path)
    display.display(img)
except FileNotFoundError:
    print(f"❌ Không tìm thấy ảnh kết quả tại {output_img_path}")
```
